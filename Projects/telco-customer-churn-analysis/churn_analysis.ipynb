## EDA Key Findings

### 1. Contract Type

Customers on month-to-month contracts have the highest churn rate at approximately **42.71%**, compared with **11.27%** for one-year contracts and **2.83%** for two-year contracts. This suggests that customers with longer-term contracts are significantly more likely to remain with the company.

### 2. Internet Service

Customers using **Fiber optic** internet have a relatively high churn rate of approximately **41.89%**, compared with **18.96%** for DSL customers and **7.40%** for customers without internet service. This indicates that fiber optic customers may require further investigation to understand the reasons behind their higher churn.

### 3. Payment Method

Customers who use **electronic checks** have the highest churn rate at approximately **45.29%**. In comparison, churn rates are much lower among customers using automatic bank transfers (**16.71%**) and automatic credit card payments (**15.24%**).

### 4. Senior Citizen Status

Senior citizens have a higher churn rate (**41.68%**) compared with non-senior customers (**23.61%**). This suggests that age-related factors may play a role in customer retention.

### 5. Customer Tenure

Customers with shorter tenure appear more likely to churn, while customers who have remained with the company for longer periods are more likely to stay. This suggests that the early stages of the customer relationship are particularly important for retention.

### 6. Monthly Charges

Customers who churned have higher average monthly charges (**approximately $74.44**) compared with customers who stayed (**approximately $61.27**). Higher monthly costs may therefore be associated with increased churn risk.

### 7. Overall Insight

The analysis suggests that customer churn is strongly associated with **contract type, tenure, internet service, payment method, monthly charges, and senior citizen status**. These factors can be explored further when building a machine learning model to predict customers who are at risk of churning.

## Business Insights and Recommendations

Based on the exploratory data analysis, several customer segments show a higher risk of churn.

### 1. Encourage Long-Term Contracts

Customers on month-to-month contracts have a significantly higher churn rate than customers on one-year or two-year contracts. The company should consider offering incentives, discounts, or additional benefits to encourage customers to move to longer-term contracts.

### 2. Improve Retention of New Customers

Customers with shorter tenure are more likely to churn. The company should strengthen its onboarding process and introduce targeted engagement campaigns during the first few months of the customer relationship.

### 3. Investigate Fiber Optic Customer Churn

Fiber optic customers have a relatively high churn rate. The company should investigate whether this is related to pricing, service quality, technical issues, or customer expectations.

### 4. Review Electronic Check Customers

Electronic-check users have the highest churn rate among payment methods. The company could investigate whether payment experience, billing issues, or customer characteristics associated with this payment method contribute to the higher churn rate.

### 5. Develop Targeted Support for Senior Customers

Senior citizens have a higher churn rate than non-senior customers. The company could explore specialized customer support, simplified service options, and personalized retention programs for this customer segment.

### 6. Review Pricing and Monthly Charges

Customers who churn tend to have higher average monthly charges. The company should investigate whether high monthly costs, pricing structure, or the combination of multiple services contributes to customer dissatisfaction.

### Overall Recommendation

The company should prioritize early customer engagement, encourage longer-term contracts, investigate the high churn among fiber optic and electronic-check customers, and develop targeted retention strategies for high-risk customer segments. These findings can also be used to develop a machine learning model that predicts customers who are most likely to churn.



```python
# Create features (X) and target (y)

X = df.drop(columns=["Churn", "customerID"])
y = df["Churn"]

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
# Convert categorical variables into numerical variables

X_encoded = pd.get_dummies(X, drop_first=True)

print("Original features:", X.shape)
print("Encoded features:", X_encoded.shape)
```

```python
# Convert churn labels to numerical values

y_encoded = y.map({
    "No": 0,
    "Yes": 1
})

print(y_encoded.value_counts())
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

print("Training features:", X_train.shape)
print("Testing features:", X_test.shape)
print("Training target:", y_train.shape)
print("Testing target:", y_test.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

print("Training features:", X_train.shape)
print("Testing features:", X_test.shape)
print("Training target:", y_train.shape)
print("Testing target:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

# Create the model
logistic_model = LogisticRegression(
    max_iter=1000,
    random_state=42
)

# Train the model
logistic_model.fit(X_train, y_train)

print("Logistic Regression model trained successfully!")
```

```python
from sklearn.linear_model import LogisticRegression

# Create the Logistic Regression model with more iterations
logistic_model = LogisticRegression(
    max_iter=5000,
    random_state=42
)

# Train the model again
logistic_model.fit(X_train, y_train)

print("Logistic Regression model trained successfully!")
```

```python
# Make predictions on the test data

y_pred = logistic_model.predict(X_test)

print("Predictions generated successfully!")
print(y_pred[:10])
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)

print(f"Model Accuracy: {accuracy:.2%}")
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Create confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display confusion matrix
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=["Stayed", "Churned"]
)

disp.plot()

plt.title("Logistic Regression Confusion Matrix")
plt.tight_layout()

plt.savefig(
    "images/logistic_regression_confusion_matrix.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
# Get predicted probabilities for churn

y_prob = logistic_model.predict_proba(X_test)[:, 1]

# Use a lower threshold of 40%
y_pred_40 = (y_prob >= 0.40).astype(int)

print(classification_report(y_test, y_pred_40))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create Random Forest model
rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight="balanced"
)

# Train the model
rf_model.fit(X_train, y_train)

print("Random Forest model trained successfully!")
```

```python
# Make predictions with Random Forest

rf_pred = rf_model.predict(X_test)

print("Random Forest predictions generated successfully!")
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Calculate accuracy
rf_accuracy = accuracy_score(y_test, rf_pred)

print(f"Random Forest Accuracy: {rf_accuracy:.2%}")
print("\nClassification Report:")
print(classification_report(y_test, rf_pred))
```

```python
# Analyze Logistic Regression coefficients

feature_importance = pd.DataFrame({
    "Feature": X_encoded.columns,
    "Coefficient": logistic_model.coef_[0]
})

# Calculate absolute coefficient strength
feature_importance["Absolute_Coefficient"] = (
    feature_importance["Coefficient"].abs()
)

# Sort by strongest influence
feature_importance = feature_importance.sort_values(
    "Absolute_Coefficient",
    ascending=False
)

feature_importance.head(15)
```

```python
# Select the 10 strongest features

top_features = feature_importance.head(10).sort_values(
    "Coefficient"
)

plt.figure(figsize=(10, 6))

plt.barh(
    top_features["Feature"],
    top_features["Coefficient"]
)

plt.title("Top Factors Associated with Customer Churn")
plt.xlabel("Logistic Regression Coefficient")
plt.ylabel("Feature")

plt.tight_layout()

plt.savefig(
    "images/top_churn_factors.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
